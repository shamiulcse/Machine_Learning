```python
#Matplotlib labes and title
#create labels for a plot
#with pyplot, you can use xlabel() and ylabel() functions to set a label for the x-axis and y-axis
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show
```

```python
#Adding Grid Lines to a plot
#We can use grid() function to add grid lines to the plot
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.title("Sports Watch Data")
plt.grid()
plt.show()
```

```python
#Matplotlib Subplot
#Display Multiple Plots
# We will use subplot() fucntion to draw multiple plots in one figure
#Plot 1:
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1) #rows,colums and index
plt.xlabel("Time")
plt.ylabel("Speed")
plt.title("First Figure")
plt.plot(x,y)
#plot 2:
a=np.array([0,1,2,3])
b=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.xlabel("Ami")
plt.ylabel(" r Tumi")
plt.title("Amader relation")
plt.plot(a,b)
plt.show()
```

```python
#Drawing 2 plots on top of each other:
#Plot 1:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,1,1)
plt.xlabel("Time")
plt.ylabel("Date")
plt.title("Hisab")
plt.plot(x,y)
#Plot 2 
a=np.array([0,1,2,3])
b=np.array([3,8,1,10])
plt.subplot(2,1,2)
plt.plot(a,b)
plt.title("Amader Relation")
plt.xlabel("Ami")
plt.ylabel("Tume")
plt.show()
```

```python
#Draw 6 plots
#Plot 1:
a=np.array([0,1,2,3])
b=np.array([3,8,1,10])
plt.subplot(2,3,1)
plt.title("Chowbila")
plt.plot(a,b)

#Plot 2:
c=np.array([4,5,6,7])
d=np.array([10,20,30,40])
plt.subplot(2,3,2)
plt.title("Voynagar")
plt.plot(c,d)

#Plot 3:
e=np.array([8,9,10,11])
f=np.array([50,60,20,10])
plt.subplot(2,3,3)
plt.title("Modhupur")
plt.plot(e,f)

#Plot 4:
g=np.array([2,3,4,5])
h=np.array([10,20,30,40])
plt.subplot(2,3,4)
plt.title("Ghiyala")
plt.plot(g,h)

#Plot 5:
i=np.array([3,4,5,6])
j=np.array([20,30,50,10])
plt.subplot(2,3,5)
plt.title("Angaru")
plt.plot(i,j)

#Plot 6:
k=np.array([7,8,9,10])
l=np.array([100,110,90,100])
plt.subplot(2,3,6)
plt.title("Manikdear")
plt.plot(k,l)

plt.suptitle("Amader Gram")
plt.show()
```

```python
#Matplotlib Scatter with scatter() Function
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,95,78,77,85,86])
plt.xlabel("How Old Car")
plt.ylabel("The Speed of the car")
plt.scatter(x,y)
plt.show()
```

```python
#Colors Change
#Dat One, The age and speed of 13 cars
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,95,78,77,85,86])
plt.xlabel("How Old Car")
plt.ylabel("The Speed of the car")
plt.scatter(x,y,color='hotpink')

#Day two, the age and speed of 15 cars:
a=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
b=np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(a,b,color='#88c999')
plt.xlabel("How old")
plt.ylabel("The Speed of the car")

plt.show()
```

```python
#Color Each Dot
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,95,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
#colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beigh","brown","gray","cyan","magenta"])
plt.xlabel("How Old Car")
plt.ylabel("The Speed of the car")
plt.scatter(x,y,c=colors)
plt.show()
```

```python
# Define x, y, and colors for each point
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
colors = ['red', 'green', 'blue', 'yellow', 'purple']

# Create a scatter plot with multiple colors
plt.scatter(x, y, c=colors)

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot with Multiple Colours!')

plt.show()
```

```python
#Matplotlib Bars:
#Creating bars with bar() function
#Draing of 4 bars
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.show()
```

```python
#Draing of 4 bars:
#Horizontal Bars:
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y)
plt.show()
```

```python
#Draing of 4 bars
#Bar Color
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y, color="blue")
plt.show()
```

```python
#Draing of 4 bars
#Bar Width
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,width=0.5)
plt.show()
```
