```python
Patient_Status=df['Patient_Status'].value_counts()
transactions=Patient_Status.index
quantity=Patient_Status.values

figure=plt.pie(df,
             values=quantity,
             names=transactions, hole=.60,
             title="Patient Status")

figure.show()               
```
